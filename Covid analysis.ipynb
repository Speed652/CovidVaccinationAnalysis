# Title: Analysis of COVID-19 Vaccination Rates in Ontario, Canada



Introduction and Motivation:
The COVID-19 pandemic has forced global efforts to vaccinate populations to stop the spread of the virus. Understanding vaccination rates helps us understand how the push towards vaccinations took place and understanding how it happened around Ontario. In this project, we analyze COVID-19 vaccination data in Ontario to gain insights into vaccination progress 
Proposed Approach:
Question: What are the trends in COVID-19 vaccination rates across different regions of Ontario, Canada?
Data Set Description:
We used the dataset from the Ontario government at (https://data.ontario.ca/dataset/covid-19-vaccine-data-in-ontario/resource/8a89caa9-511c-4568-af89-7f2174b4378c) containing COVID-19 vaccination information. The dataset includes vaccination metrics such as the number of doses administered, and vaccinations received.
Data Processing:
Data Cleaning: The dataset was cleaned of missing values, inconsistencies, and errors.
Data Visualization: The data was visualized using appropriate graphs."


Visualizations:
Time Series Plot: Daily Change in Total Doses Administered
Our time series plot showcases when most of the vaccinations took place and pinpoints exactly when the movement toward vaccination occurred as  a society
This Time Series shows 2 main spikes in where mass vaccinations happened with the first spike being the release of the first vaccination shot, and the second spike being the second vaccine, with all the other mini spikes likely to be booster shots. 


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the vaccination data
vaccination_data = pd.read_csv('CovidData.csv')

# Data Cleaning
vaccination_data = vaccination_data[['report_date', 'total_doses_administered']]
vaccination_data['report_date'] = pd.to_datetime(vaccination_data['report_date'])

# Calculate daily change
vaccination_data['daily_change'] = vaccination_data['total_doses_administered'].diff()


# Time Series Plot
plt.figure(figsize=(12, 6))
plt.plot(vaccination_data['report_date'], vaccination_data['daily_change'], marker='o', color='blue')
plt.title('Daily Change in Vaccination Rates')
plt.xlabel('Date')
plt.ylabel('Daily Change in Total Doses Administered')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Time series: Cumulative COVID-19 Vaccinations Over Time in Ontario
This visualization shows the cumulative number of vaccinations given out in Ontario over time, with more steep areas of the graph representing more aggressive vaccination time periods
This graph indicates that the total number of shots administered is over 41 million. With Ontario's population being around 15 million, and most people taking at least 2 shots, the data indicates that majority of Ontario has been vaccinated.


```python
plt.figure(figsize=(12, 6))
plt.plot(vaccination_data['report_date'], vaccination_data['total_doses_administered'], marker='o', linestyle='-', color='blue')
plt.title('Cumulative COVID-19 Vaccinations Over Time in Ontario')
plt.xlabel('Date')
plt.ylabel('Total Doses Administered')
plt.grid(True)
plt.xticks(rotation=45)
plt.ticklabel_format(style='plain', axis='y')  
plt.tight_layout()  
plt.show()
```

Bar Chart: Proportion of Fully vs Partially Vaccinated Individuals
This visualization will compare the proportions of vaccinated people who are fully vaccinated vs partially vaccinated 


The data shows how most of the population decided to take the health officials advice and take their booster shots.


```python
vaccination_data = pd.read_csv('CovidData.csv')

vaccination_data['report_date'] = pd.to_datetime(vaccination_data['report_date'])
vaccination_data.sort_values('report_date', inplace=True)

# Extract the data
latest_data = vaccination_data.iloc[-1]
total_fully_vaccinated = latest_data['total_individuals_fully_vaccinated']
total_partially_vaccinated = latest_data['total_individuals_partially_vaccinated']

# Calculate proportions
total_vaccinated = total_fully_vaccinated + total_partially_vaccinated
proportions = [total_fully_vaccinated / total_vaccinated, total_partially_vaccinated / total_vaccinated]
labels = ['Fully Vaccinated', 'Partially Vaccinated']
colors = ['green', 'orange']

# pie chart
plt.figure(figsize=(8, 8))
plt.pie(proportions, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Fully vs Partially Vaccinated Individuals')
plt.show()
```
